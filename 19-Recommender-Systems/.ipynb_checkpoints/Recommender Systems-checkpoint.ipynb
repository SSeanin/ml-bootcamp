```python
import numpy as np
import pandas as pd
```

```python
columns_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data', names=columns_names, sep='\t')
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_palette('viridis')
```

```python
df.head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head(5)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean(), columns=['rating'])
```

```python
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
sns.displot(data=ratings, x='num of ratings', height=8)
```

```python
sns.displot(data=ratings, x='rating', bins=70, height=8)
```

```python
sns.jointplot(x='rating', y='num of ratings', data=ratings, height=8)
```

```python
user_ratings = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
user_ratings
```

```python
ratings.sort_values('num of ratings', ascending=False).head(10)
```

```python
starwars_ratings = user_ratings['Star Wars (1977)']
liarliar_ratings = user_ratings['Liar Liar (1997)']
```

```python
similar_to_starwars = user_ratings.corrwith(starwars_ratings)
similar_to_liarliar = user_ratings.corrwith(liarliar_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])

corr_starwars.dropna(inplace=True)
corr_liarliar.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars[corr_starwars['num of ratings'] > 150].sort_values('Correlation', ascending=False).head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
```

```python
corr_liarliar[corr_liarliar['num of ratings'] > 100].sort_values('Correlation', ascending=False).head()
```
